```python
def getCorrelatedFeature(corrdata, threshold):
    feature = []
    value = []
    
    for i, index in enumerate(corrdata.index):
        if abs(corrdata[index])> threshold:
            feature.append(index)
            value.append(corrdata[index])

    trade_data = pd.DataFrame(data = value, index = feature, columns=['Corr Value'])
    return trade_data
threshold = 0.4
corr_value = getCorrelatedFeature(trade_data.corr()['Cumulative'], threshold)
```

```python
corr_value.index.values
```

```python
correlated_data = trade_data[corr_value.index]
correlated_data.head()
```

```python
import numpy as np

# plot a bar plot for each categorical feature count
categorical_features = ['Year', 'Weekday', 'Commodity']

for col in categorical_features:
    counts = trade_data[col].value_counts().sort_index()
    fig = plt.figure(figsize=(9, 6))
    ax = fig.gca()
    counts.plot.bar(ax = ax, color='lightblue')
    ax.set_title(col + ' counts')
    ax.set_xlabel(col) 
    ax.set_ylabel("Frequency")
plt.show()
```
